```python
import pandas as pd
df = pd.read_csv("/content/data.csv?1770634820")
```

```python
df
```

```python
products= df[['product_id', 'product_name', 'category', 'discounted_price', 'actual_price','calculated_discount']].drop_duplicates()
```

```python
products
```

```python
#PANDAS MELT


new_products= pd.melt(products,
        id_vars= ['product_id', 'product_name', 'category'],
        value_vars=['discounted_price', 'actual_price','calculated_discount'],
        var_name='col',
        value_name='value')

```

```python
products_reversal= pd.pivot_table(new_products,
               index=['product_id', 'product_name', 'category'],
               columns= 'col',
               values= 'value')
```
